```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import matplotlib as mpl

mpl.rcParams['font.family'] = 'IBM Plex Sans KR'
mpl.rcParams['axes.unicode_minus'] = False
mpl.rcParams['font.size'] = 11
```

### 히스토그램
* 도수분포표(빈도표)를 시각화한 것
* bins : 구간 갯수 지정

```python
weights = [68, 81, 64, 56, 78, 74, 61, 77, 66, 68, 59, 71,
          80, 59, 67, 81, 69, 73, 69, 74, 70, 65]
np.min(weights), np.max(weights)
```

```python
plt.hist(weights, color='c')
plt.tight_layout()
```

```python
# 5개 구간으로 나누기
plt.hist(weights, color='c', bins=5)
plt.tight_layout()
```

```python
plt.hist(weights, color='b', bins=10)
plt.tight_layout()
```

```python
### 확률분포 표시 KDE
plt.hist(weights, color='c', bins=10, density=True, alpha=0.7)
plt.tight_layout()
```

### 박스 플롯
* 최소값, 1사분위수, 2사분위수, 3사분위수, 최대값 등의 수치를 시각화

```python
ax = plt.boxplot(weights)
plt.tight_layout()
```

```python
# 최대값, 최소값
[item.get_ydata() for item in ax['whiskers']]
```

```python
# 중앙값
[item.get_ydata() for item in ax['medians']]
```

```python
# 이상치 (수염 밖에 있는 값): IQR (3사분위수 - 1사분위 미만/초과)
[item.get_ydata() for item in ax['fliers']]
```

### 파이그래프

```python
ratio = [34, 32, 16, 18]
vendor = ['Apple','Saumsung','Motolora','ETC']
```

```python
plt.pie(ratio, labels=vendor, autopct='%.0f%%')
plt.tight_layout()
```

```python
plt.pie(ratio, labels=vendor, autopct='%.0f%%', startangle=90, counterclock=False)
plt.tight_layout()
```

```python
# 중심에서 멀어진 정도
explode = [0.05, 0.05, 0.05, 0.05]
plt.pie(ratio, labels=vendor, autopct='%.0f%%', startangle=90, counterclock=False, explode=explode, shadow=True)
plt.tight_layout()
```

```python
colors = ['cyan','orange','yellow','green']
plt.pie(ratio, labels=vendor, autopct='%.0f%%', startangle=90, counterclock=False, colors=colors)
plt.tight_layout()
```

```python
colors = plt.cm.Set2(range(len(vendor)))
plt.pie(ratio, labels=vendor, autopct='%.0f%%', startangle=90, counterclock=False, colors=colors)
plt.tight_layout()
```

```python
edgestyle = {'width':0.7, 'edgecolor':'w', 'linewidth':5}
plt.pie(ratio, labels=vendor, autopct='%.0f%%', startangle=90, counterclock=False, colors=colors, wedgeprops=edgestyle)
plt.tight_layout()
```
