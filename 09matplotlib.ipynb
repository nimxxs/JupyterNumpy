## matplotlib
* 파이썬의 대표적인 시각화도구

#### 선그래프 : plot
* 정한 자료는 자동적으로 y축으로 설정
* x축 값이 없으면 자동으로 0,1,2,3,4, ... 로 설정

```python
data1 = pd.Series([1,4,9,16,25])
plt.plot(data1)
plt.show()
```

```python
# x축 값 지정
np.random.seed(2309131155)
y = np.random.standard_normal(20)  # 표준정규분포를 따르는 난수 20개
x = np.arange(1,20+1)
```

```python
plt.plot(y)  # x축 자동감지
plt.show()  
```

```python
plt.plot(x, y)  # x축 자동감지
plt.show()  
```

#### 암시적 스타일 지정 : 색
* r, g, b, c, m, y, k, w

```python
x = [1,2,3,4,5]
y = [9,8,7,6,5]
plt.plot(x, y, 'c')
plt.show()
```

#### 암시적 스타일 지정 : 마커
* 데이터 위치를 나타내는 기호
* . / o / v / ^ / 1 / p / * / D

```python
plt.plot(x, y, 'c*')
plt.show()
```

#### 암시적 스타일 지정 : 선종류
* solid, dashed, dotted, dahs-dot 등이 있음
* \- / : / -. / --

```python
plt.plot(x, y, 'c-.')
plt.show()
```

# 명시적 스타일 지정
* color, linewidth, linestyle, markersize,
* marker edge color, marker edge width, 
* marker face color

```python
# plt.plot(x, y, color='red')
plt.plot(x, y, c='c', linewidth=3)
plt.show()
```

```python
plt.plot(x, y, c='m', lw=3, ls=':', marker='*',
        ms='20', mec='c', mew=3, mfc='y')
plt.show()
```

### 축 범위지정
* xlim(최소, 최대), ylim(최소, 최대)로 설정

```python
plt.plot(x, y)
plt.xlim(-1, 6)
plt.ylim(3, 15)
plt.show()
```

#### 축 눈금 지정하기
* 축 상의 위치 표시지점을 tick이라 함
* xticks(수치리스트, 레이블)
* yticks(수치리스트, 레이블)

```python
x = np.linspace(-np.pi, np.pi, 256)
y = np.cos(x)
plt.plot(x, y)
plt.show()
# x축 눈금은 다소 뜨문뜨문 표시
# y축 눈금은 다소 촘촘하게 표시
```

```python
plt.plot(x, y)
plt.yticks([-1, -0.5, 0, 0.5, 1])
plt.xticks(np.linspace(-np.pi, np.pi, 6))
plt.show()
```

#### grid line : 안내선
* 틱의 위치를 잘 보여주기 위해 사용
* grid(True/False)

```python
plt.plot(x, y)
plt.grid(True)
plt.show()
```

# 범례 추가하기 : legend
* 여러 개의 plot을 동시에 그리는 경우에
* 각 plot이 어떤 자료인지 나타내기 위해 사용
* 먼저, 각 그래프에 label 속성으로 이름을 작성해둬야 함

```python
x = np.linspace(-3.14, 3.14, 256)
y1 = np.cos(x)
y2 = np.sin(x)
```

```python
plt.plot(x, y1, 'm--', label='cosine')
plt.plot(x, y2, 'c--', label='sine')
plt.legend(loc='best')
plt.show()
```
