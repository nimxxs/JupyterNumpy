### seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### 선그래프

```python
x = np.linspace(0, 14, 100)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
```

```python
sns.lineplot(x=x, y=y1, color='c', label='sine')
sns.lineplot(x=x, y=y2, color='y', label='cosine')
sns.lineplot(x=x, y=y3, color='m', label='tan')
plt.tight_layout()
```

### 월별 항공기 탑승객 데이터 시각화

```python
df = sns.load_dataset('flights')
df.head()
```

```python
# 상한/하한 표시
sns.lineplot(x='year', y='passengers', data=df)
plt.tight_layout()
```

```python
sns.lineplot(x='year', y='passengers', data=df, hue='month')
plt.tight_layout()
```

```python
# 범주 제한 : hut_order
order = ['Jan','Feb','Mar']
sns.lineplot(x='year', y='passengers', data=df, hue='month', hue_order=order)
plt.tight_layout()
```

### 팁 데이터 시각화

```python
tips = sns.load_dataset('tips')
```

```python
find = tips.day == 'Fri'
tips[find]['total_bill'].mean()
```

```python
# 요일별 평균 total_bill 시각화
sns.barplot(y='day', x='total_bill', data=tips, errwidth=False, orient='h')
plt.tight_layout()
```

```python
cm = sns.color_palette("Set2")

sns.barplot(y='day', x='total_bill', data=tips, errwidth=False, orient='h', palette=cm)
plt.tight_layout()
```

```python
# 요일별 성별 평균 total_bill 시각화
sns.barplot(x='day', y='total_bill', data=tips, hue='sex', errwidth=False)
plt.tight_layout()
```

```python
ax = sns.barplot(x='day', y='total_bill', data=tips, hue='sex', errwidth=False)

for p in ax.patches:
    plt.annotate(p.get_height().round(2), (p.get_x()+0.06, p.get_height()+0.3), color='navy')
    
plt.tight_layout()
```

```python
# estimator : mean, sum, min, max, size, median
ax = sns.barplot(x='day', y='total_bill', data=tips, hue='sex', errwidth=False, estimator=np.median)

for p in ax.patches:
    plt.annotate(p.get_height().round(2), (p.get_x()+0.08, p.get_height()+0.3), color='navy')
    
plt.tight_layout()
```

### 산점도와 회귀선 그래프

```python
import warnings
warnings.simplefilter(action='ignore', category=UserWarning)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
plt.tight_layout()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, line_kws={'color':'red', 'linestyle':'--',})
plt.tight_layout()
```

```python
# ci : confidence interval 신뢰구간
sns.lmplot(x='total_bill', y='tip', data=tips, line_kws={'color':'red', 'linestyle':'--',}, ci=False)
plt.tight_layout()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, ci=False, hue='smoker', markers=['*','o'])
plt.tight_layout()
```
